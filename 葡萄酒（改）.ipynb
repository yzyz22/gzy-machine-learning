手动设置数据集划分结果 修改神经网络层数 

数据集手动划分为70%训练集 20%验证集 10%测试集

```python
import pandas as pd
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# 导入数据
raw_data = pd.read_csv('wine.data', header=None)

# 样本数
data_num = raw_data.shape[0]
```

手动设置数据集划分索引

```python
train_indexes = np.arange(0, int(data_num * 0.7))  # 训练集前70%
val_indexes = np.arange(int(data_num * 0.7), int(data_num * 0.9))  # 验证集20%
test_indexes = np.arange(int(data_num * 0.9), data_num)  # 测试集10%

train_data = raw_data.loc[train_indexes]  # 训练集
val_data = raw_data.loc[val_indexes]  # 验证集
test_data = raw_data.loc[test_indexes]  # 测试集
```

三份数据进行合并

```python
train_validation_data = pd.concat([train_data, val_data, test_data])
mean = train_validation_data.mean()  # 均值
std = train_validation_data.std()  # 标准差
```

```python
train_data = (train_data - mean) / std
val_data = (val_data - mean) / std
test_data = (test_data - mean) / std
```

```python
# 转换为numpy的array类型
x_train = np.array(train_data.iloc[:, 1:])
y_train = np.array(train_data.iloc[:, 0])
x_val = np.array(val_data.iloc[:, 1:])
y_val = np.array(val_data.iloc[:, 0])
x_test = np.array(test_data.iloc[:, 1:])
y_test = np.array(test_data.iloc[:, 0])
```

神经网络层数设置为6

```python
model = keras.Sequential([
    keras.Input(shape=(13,)),  # 13个输入
    layers.Dense(12, activation='relu'),  # 第一层12个节点
    layers.Dense(16, activation='relu'),  # 第二层16个节点
    layers.Dense(8, activation='relu'),  # 第三层8个节点
    layers.Dense(4, activation='relu'),  # 第四层4个节点
    layers.Dense(2, activation='relu'),  # 第五层2个节点
    layers.Dense(1,)  # 第六层1个节点，即一个输出
])
```

```python
# 查看模型的信息
model.summary()
```

指定训练配置（优化器为sgd、损失函数为平均绝对误差、评估指标）

```python
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.01),
    loss=keras.losses.MeanAbsoluteError(),
    metrics=[keras.metrics.MeanAbsoluteError()]  #绝对误差
)
```

```python
# 模型训练
history = model.fit(
    x_train, y_train,
    batch_size=64,
    epochs=500,
    validation_data=(x_val, y_val)
)
```
